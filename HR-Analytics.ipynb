```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_csv('HR-Employee-Attrition.csv')
df
```

# Distribution of Employees Across Departments

```python
# Plot
sns.countplot(x='Department', data=df)
plt.title('Distribution of Employees Across Departments')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.show()
```

# Gender Distribution

```python
sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Number of Employees')
plt.show()
```

# Age Distribution

```python
sns.histplot(df['Age'], bins=10)
plt.title('Age Distribution of Employees')
plt.xlabel('Age')
plt.ylabel('Number of Employees')
plt.show()
```

# Overall Attrition Rate

```python
attrition_rate = df['Attrition'].value_counts(normalize=True)['Yes'] * 100
print(f'Overall Attrition Rate: {attrition_rate:.2f}%')
```

# Attrition Rate by Department

```python
department_attrition = df[df['Attrition'] == 'Yes']['Department'].value_counts(normalize=True) * 100
sns.barplot(x=department_attrition.index, y=department_attrition.values)
plt.title('Attrition Rate by Department')
plt.xlabel('Department')
plt.ylabel('Attrition Rate (%)')
plt.show()
```

# Attrition Rate by Age Group

```python
df['AgeGroup'] = pd.cut(df['Age'], bins=[20, 30, 40, 50, 60], labels=['20-30', '30-40', '40-50', '50-60'])
age_attrition = df[df['Attrition'] == 'Yes']['AgeGroup'].value_counts(normalize=True) * 100
sns.barplot(x=age_attrition.index, y=age_attrition.values)
plt.title('Attrition Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Attrition Rate (%)')
plt.show()
```

# Monthly Income Distribution

```python
sns.boxplot(x='Department', y='MonthlyIncome', data=df)
plt.title('Monthly Income Distribution Across Departments')
plt.xlabel('Department')
plt.ylabel('Monthly Income')
plt.show()
```

# Monthly Income by Job Role

```python
sns.boxplot(x='JobRole', y='MonthlyIncome', data=df)
plt.title('Monthly Income by Job Role')
plt.xlabel('Job Role')
plt.ylabel('Monthly Income')
plt.xticks(rotation=90)
plt.show()
```

# Correlation Between Monthly Income and Years at Company

```python
sns.scatterplot(x='YearsAtCompany', y='MonthlyIncome', data=df)
plt.title('Correlation Between Monthly Income and Years at Company')
plt.xlabel('Years at Company')
plt.ylabel('Monthly Income')
plt.show()
```

# Job Satisfaction Distribution

```python
sns.countplot(x='JobSatisfaction', data=df)
plt.title('Job Satisfaction Distribution')
plt.xlabel('Job Satisfaction Level')
plt.ylabel('Number of Employees')
plt.show()
```

# Job Satisfaction by Department

```python
department_satisfaction = df.groupby('Department')['JobSatisfaction'].mean().sort_values()
sns.barplot(x=department_satisfaction.index, y=department_satisfaction.values)
plt.title('Average Job Satisfaction by Department')
plt.xlabel('Department')
plt.ylabel('Average Job Satisfaction')
plt.show()
```

# Correlation Between Job Satisfaction and Attrition

```python
satisfaction_attrition = df[df['Attrition'] == 'Yes']['JobSatisfaction'].value_counts(normalize=True) * 100
sns.barplot(x=satisfaction_attrition.index, y=satisfaction_attrition.values)
plt.title('Attrition Rate by Job Satisfaction')
plt.xlabel('Job Satisfaction Level')
plt.ylabel('Attrition Rate (%)')
plt.show()
```

# Average Number of Trainings by Department

```python
training_by_department = df.groupby('Department')['TrainingTimesLastYear'].mean()
sns.barplot(x=training_by_department.index, y=training_by_department.values)
plt.title('Average Number of Trainings by Department')
plt.xlabel('Department')
plt.ylabel('Average Number of Trainings')
plt.show()
```

# Training Frequency and Performance Ratings

```python
sns.scatterplot(x='TrainingTimesLastYear', y='PerformanceRating', data=df)
plt.title('Training Frequency and Performance Ratings')
plt.xlabel('Training Times Last Year')
plt.ylabel('Performance Rating')
plt.show()
```

# Correlation Between Training Frequency and Attrition

```python
training_attrition = df[df['Attrition'] == 'Yes']['TrainingTimesLastYear'].value_counts(normalize=True) * 100
sns.barplot(x=training_attrition.index, y=training_attrition.values)
plt.title('Attrition Rate by Training Frequency')
plt.xlabel('Training Times Last Year')
plt.ylabel('Attrition Rate (%)')
plt.show()
```

# Work-Life Balance Distribution

```python
sns.countplot(x='WorkLifeBalance', data=df)
plt.title('Work-Life Balance Distribution')
plt.xlabel('Work-Life Balance Rating')
plt.ylabel('Number of Employees')
plt.show()
```

# Work-Life Balance by Job Role

```python
worklife_by_jobrole = df.groupby('JobRole')['WorkLifeBalance'].mean()
sns.barplot(x=worklife_by_jobrole.index, y=worklife_by_jobrole.values)
plt.title('Work-Life Balance by Job Role')
plt.xlabel('Job Role')
plt.ylabel('Average Work-Life Balance Rating')
plt.xticks(rotation=90)
plt.show()
```

# Correlation Between Work-Life Balance and Attrition

```python
worklife_attrition = df[df['Attrition'] == 'Yes']['WorkLifeBalance'].value_counts(normalize=True) * 100
sns.barplot(x=worklife_attrition.index, y=worklife_attrition.values)
plt.title('Attrition Rate by Work-Life Balance')
plt.xlabel('Work-Life Balance Rating')
plt.ylabel('Attrition Rate (%)')
plt.show()
```
